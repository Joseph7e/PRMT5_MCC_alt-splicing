```python
import pandas as pd
import matplotlib.pyplot as plt


files = ['../input-files/subsample.txt']

for file_path in files:
    # Load the TSV file into a DataFrame
    #file_path = 'jc-1.subsampling_report.txt'  # Replace with the path to your TSV file
    data = pd.read_csv(file_path, sep='\t')

    # Display the available columns to choose from
    print("Available columns:", data.columns)

    # Select the column you want to plot for the saturation curve
    # Replace 'unique_genes' with the column name you want to plot
    column_to_plot = 'unique_genes'
    column_to_plot = 'unique_isoforms'

    # Plot the saturation curve
    plt.figure(figsize=(10, 6))
    plt.plot(data['reads'], data[column_to_plot], marker='o', linestyle='-', color='b', label=column_to_plot)

    # Adding labels and title
    plt.xlabel('Number of Reads')
    plt.ylabel(column_to_plot.replace('_', ' ').capitalize())
    plt.title(f'Saturation Curve: {column_to_plot.replace("_", " ").capitalize()}')
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()
```
